```python
class Solution(object):
    def validPath(self, n, edges, source, destination):
        """
        :type n: int
        :type edges: List[List[int]]
        :type source: int
        :type destination: int
        :rtype: bool
        """
        adjacency_list = [[] for _ in range(n)] # makes adjacencey list from given edges
        for u, v in edges:
            adjacency_list[u].append(v)
            adjacency_list[v].append(u)

        visited = set() # keep track of visited nodes during traversal/recursion

        def dfs(node, visited):
            if node == destination: # base case
                return True
            visited.add(node)
            for neighbor in adjacency_list[node]: # for each neighbor node
                if neighbor not in visited: # if it has not been visited
                    if dfs(neighbor, visited): # call dfs on that neighbor
                        return True
            return False # if a path to destination was not found
    
        return dfs(source, visited) # starting dfs from the source node

solution = Solution()
n = 3
edges = [[0,1],[1,2],[2,0]]
source = 0
destination = 2
print(solution.validPath(n, edges, source, destination))
```
