```python
def is_looping(n):
    def next_number(number):
        next = 0
        while number > 0:
            current = number % 10 # get the last digit
            next += current**2 # get the last digits square
            number //= 10 # remove last digit
        return next
        
    visited = set()
    while n != 1:
        if n in visited: # check if n has already been in the sequence
            return True # if so its looping
        visited.add(n)
        n = next_number(n) # else check next n
    return False

print(is_looping(4))
print(is_looping(19))
```
