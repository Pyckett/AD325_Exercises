data lemur problem:

```python
def is_looping(n):
    def next_number(number):
        next = 0
        while number > 0:
            current = number % 10 # get the last digit
            next += current**2 # get the last digits square
            number //= 10 # remove last digit
        return next
        
    visited = set()
    while n != 1:
        if n in visited: # check if n has already been in the sequence
            return True # if so its looping
        visited.add(n)
        n = next_number(n) # else check next n
    return False

print(is_looping(4))
print(is_looping(19))
```

leet code problem:

```python
class Solution(object):
    def maxSubsequence(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[int]
        """
        nums_with_indices = [(num, i) for i, num in enumerate(nums)] # pair each number with its index
        nums_with_indices.sort(key=lambda x: -x[0]) # sort by value descending
        top_k = sorted(nums_with_indices[:k], key=lambda x: x[1]) # sort k elements by their index
        return [num for num, _ in top_k] # get the values

solution = Solution()
numbers = [2, 1, 3, 3]
print(solution.maxSubsequence(numbers, 2))
```
