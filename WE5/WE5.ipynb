data lemur problem:

```python
def max_avg_subarray(nums, k):
  max_sum = float("-inf")
  for i in range(len(nums)):
    if i + k <= len(nums): # checks if there is a subarray within bounds i to i + k
      current_sum = 0
      for j in range(i, i + k):
        current_sum += nums[j] # sums the next k number of values
      max_sum = max(max_sum, current_sum) # updates max_sum if current subaray sum is larger
  return round(max_sum / k, 2)
```

leet code problem:

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def insertIntoBST(self, root, val):
        if root is None: # empty tree base case
            return TreeNode(val) 
        if root.val > val: # value recursively inserted into left subtree
            root.left = self.insertIntoBST(root.left, val)
        else: # value recursively inserted to right subtree
            root.right = self.insertIntoBST(root.right, val)
        return root
```
