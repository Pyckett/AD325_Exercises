data lemur problem:

```python
def max_three(input):
    max_triplet = input[0] * input[1] * input[2]  # establish our first triplet for comparisons
    for i in range(len(input) - 2):
        for j in range(i + 1, len(input) - 1):
            for k in range(j + 1, len(input)):
                current_triplet = input[i] * input[j] * input[k]  # this loop checks other triplet's by setting two constants at a time and one variable to make each new current product.
                if current_triplet > max_triplet:
                    max_triplet = current_triplet
    return max_triplet
```

code step by step problem:

```python
class Solution:
    def isBalanced(self, root):
        def dfs(root): # using a depth first search traversal method
            if not root:
                return (True, 0) # null tree or subtree encountered

            left_balanced, left_height = dfs(root.left) # rescursion through left subtrees
            right_balanced, right_height = dfs(root.right) # recursion through right subtrees

            is_balanced = left_balanced and right_balanced and abs(left_height - right_height) <= 1 # checks on our recurison results if left and right subtrees are all balanced and checks the balance factor using absolute value so we focus on the difference in height.
            return (is_balanced, 1 + max(left_height, right_height)) # returns if it is balanced and the height

        return dfs(root)[0]
```
